```python
from selenium import webdriver
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
from urllib.parse import quote_plus
from selenium.webdriver.common.keys import Keys
import time
import requests
import pandas as pd

url = 'https://gall.dcinside.com/board/lists?id=dcbest'
response = requests.get(url, headers={'User-agent': 'Mozilla/5.0'})

html = response.text
soup = bs(html, 'lxml')
article=soup.select('#container > section.left_content > article')
tr=article[1].select('tr')
```

```python
import warnings
warnings.simplefilter(action='ignore')
```

```python
urlbase='https://gall.dcinside.com/'
l=[]
for i in range(3,52):
  url2=urlbase+tr[i].select('a')[0]['href']
  response2 = requests.get(url2, headers={'User-agent': 'Mozilla/5.0'})
  html2 = response2.text
  soup2 = bs(html2, 'lxml')
  chrome_options = webdriver.ChromeOptions()
  chrome_options.add_argument('--headless')
  chrome_options.add_argument('--no-sandbox')
  chrome_options.add_argument('--disable-dev-shm-usage')
  driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)
  driver.get(url2)
  time.sleep(2)
  ul = driver.find_element_by_css_selector('#focus_cmt>div > div.comment_box > ul')
  li = ul.find_elements_by_tag_name("li")
  
  for index, value in enumerate(li):
      re=value.find_elements_by_css_selector("div > div.clear.cmt_txtbox.btn_reply_write_all")
      for index, value in enumerate(re):
        try:
          p=value.find_element_by_tag_name('p').text
        except:
          p=''
        l.append(p)
  print('============================================={}==================================='.format(i))
```

```python
j=2
while j<15:
  url = 'https://gall.dcinside.com/board/lists?id=dcbest'
  url+='&page={}'.format(j)
  response = requests.get(url, headers={'User-agent': 'Mozilla/5.0'})
  html = response.text
  soup = bs(html, 'lxml')
  article=soup.select('#container > section.left_content > article')
  tr=article[1].select('tr')
  j+=1
  for i in range(2,52):
    url2=urlbase+tr[i].select('a')[0]['href']
    response2 = requests.get(url2, headers={'User-agent': 'Mozilla/5.0'})
    html2 = response2.text
    soup2 = bs(html2, 'lxml')
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)
    driver.get(url2)
    time.sleep(2)
    ul = driver.find_element_by_css_selector('#focus_cmt>div  > div.comment_box > ul')
    li = ul.find_elements_by_tag_name("li")
    for index, value in enumerate(li):
        re=value.find_elements_by_css_selector("div > div.clear.cmt_txtbox.btn_reply_write_all")
        for index, value in enumerate(re):
          try:
            p=value.find_element_by_tag_name('p').text
          except:
            continue
          l.append(p)
    print('============================================={}==================================='.format(i))
```
