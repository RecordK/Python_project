```python
import pandas as pd
df=pd.read_csv('/content/possum.csv')
```

```python
df.isna().sum()
```

```python
df=pd.get_dummies(df)
df.drop(['Pop_other','sex_f'],axis=1,inplace=True)
```

```python
df['age']=df['age'].fillna(df['age'].mean())
```

```python
df['footlgth']=df['footlgth'].fillna(df['footlgth'].mean())
```

```python
import copy
df1=copy.deepcopy(df)
df1.drop('age',axis=1,inplace=True)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# 데이터
x_data=df1
y_data=df[['age']]
x_train = torch.FloatTensor(x_data.values)
x_train = F.normalize(x_train, dim=1)
y_train = torch.FloatTensor(y_data.values)
y_train
```

```python
model = nn.Linear(13,1)  #리니어 모델 생성

# 3. 비용 최적화할 옵티마이저 생성. 
optimizer = torch.optim.SGD(model.parameters(), lr=0.1) 

# 4. 학습 진행
nb_epochs = 10000
for epoch in range(nb_epochs+1):
 
    prediction = model(x_train)  #현재 모델의 가중치와 편향으로 예측
  
    cost = F.mse_loss(prediction, y_train) #예측된 값과 정답 사이의 비용 계산


    optimizer.zero_grad()  
    cost.backward()
    optimizer.step()

    if epoch % 1000 == 0:

          print('Epoch {:4d}/{} Cost: {:.6f}'.format(
          epoch, nb_epochs, cost.item()
      ))
```

```python
pred=model(x_train)
print(pred,y_train)
```
