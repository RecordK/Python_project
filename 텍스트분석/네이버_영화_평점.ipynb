```python
from selenium import webdriver
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
from urllib.parse import quote_plus
from selenium.webdriver.common.keys import Keys
import time
import requests
import pandas as pd
url = 'https://www.kobis.or.kr/kobis/business/stat/boxs/findRealTicketList.do'

# 크롬 옵션 설치하기
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("user-agent=Mozilla/5.0")
driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)

# 웹페이지 이동
driver.get('https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=190991&target=')

# 조회버튼 요소 검색해서 클릭
# driver.find_element_by_xpath('//*[@id="searchForm"]/div/div[4]/button[1]').click()
dataset=[]
title = driver.find_element_by_xpath("//*[@id='old_content']/table/tbody/tr[1]/td[2]/a[1]").text
avg_score = driver.find_element_by_css_selector('#old_content > div.choice_movie_box > div.choice_movie_info > table > tbody > tr:nth-child(7) > td > strong').text
def review():
  title = driver.find_element_by_xpath("//*[@id='old_content']/table/tbody/tr[1]/td[2]/a[1]").text
  # div = driver.find_element_by_css_selector('div.choice_movie_info')
  # title = div.find_element_by_tag_name("h5").find_element_by_tag_name("a").text
  avg_score = driver.find_element_by_css_selector('#old_content > div.choice_movie_box > div.choice_movie_info > table > tbody > tr:nth-child(7) > td > strong').text
  content=[]
  for i in range(1,11):
    content.append(driver.find_element_by_xpath("//*[@id='old_content']/table/tbody/tr[{}]/td[2]".format(i)).text)
  wridate=[]
  for i in range(1,11):
    wridate.append(driver.find_element_by_xpath("//*[@id='old_content']/table/tbody/tr[{}]/td[3]".format(i)).text)
  print(title)
  print(avg_score)
  print('==============')
  for i,j in zip(content,wridate):
    title1=i.split('\n')[0]
    print(title1)
    star=i.split('\n')[2]
    print(star)
    rev=i.split('\n')[3].split(' 신고')[0]
    print(rev)
    id=j.split('\n')[0]
    date=j.split('\n')[1]
    print(id)
    print(date)
    print('==============')
    dataset.append([title1,star,rev,id,date])
for i in range(1,11):
  driver.find_element_by_xpath('//*[@id="old_content"]/div[2]/div/a[{}]/span'.format(i)).click()

  time.sleep(3)
  review()
  print('=====================================',i,'페이지=========================')
```

```python
import pandas as pd
# sheet1_cols = ['review','score','writer']

cols=['title','star', 'review','id','date']
df = pd.DataFrame(dataset, columns=cols)
df.loc[df['review']=='신고','review']='없음'
df
```
