```python
import pandas as pd
df= pd.read_csv('/content/DBPEDIA_test.csv')
```

```python
x_data=df['text']
```

```python
y_data=df['l1']
cate=y_data.unique()
```

```python
import re
x_data=x_data.apply(lambda x:re.sub('[^a-zA-Z]',' ',x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect= CountVectorizer(max_df=0.95, max_features=1000,min_df=2,stop_words='english',ngram_range=(1,2))
feat_vect = count_vect.fit_transform(x_data)
```

```python
feature_names=count_vect.get_feature_names()
feature_names
```

```python
from sklearn.decomposition import LatentDirichletAllocation

# 토픽 추출 모델 객체 생성 n_components 값은 주제의 개수
# 토픽을 추출하기 위해 n_components 값 만큼 글의 카테고리 분류류
# 글들을 주제별로 분류한 뒤 그 글들에서 단어들의 중요도를 계산
lda = LatentDirichletAllocation(n_components=len(cate), random_state=0)
lda.fit(feat_vect)
```

```python
import copy
print(lda.components_.shape)
lda.components_
```

```python
a=copy.deepcopy(lda.components_)
```

```python
a[0].argsort()[::-1][:10]
```

```python
for idx,i in enumerate(a):
  print('==topic',idx,'=======')
  sorted_idx=i.argsort()[::-1][:10]
  fe_concat=' '.join([feature_names[l]for l in sorted_idx])
  print('주요단어 :',fe_concat)
```

```python
for idx,i in enumerate(a):
  print('==topic',idx,'=======')
  s=''
  sorted_idx=i.argsort()[::-1][:10]
  for j in sorted_idx:
    s+=feature_names[j]+'\t'
  print('주요단어 :',s)
```
